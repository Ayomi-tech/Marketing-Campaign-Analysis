Since the missing data is 24 out of 2,240 which is 1.07% of the total data, we can drop these missing values without signigicant effect on the integrity of the data.


```python
# Dropping missing values

df = df.dropna()
```

```python
# Extracting the unique values from Education and Marital_status fields


print("Unique values in Education field:" , df['Education'].unique())
print("Unique values in Marital_Status field:" , df['Marital_Status'].unique())
```

```python
# Create "Education_level" field base on the following condition

def education_level(education):
    
    if education in ['Graduation', 'PhD', 'Master']:
        
        return "High"
    
    elif education in ['Basic']:
        
        return "Middle"
    
    else:
        return "Low"
    
    
df['Education_level'] = df['Education'].apply(education_level)
```

```python
# Create "Living_Status" field base on the following condition

def living_status(marital_status):
    
    if marital_status in ['Alone', 'Absurd', 'YOLO']:
        
        return "Living Alone"
    
    else:
        return "Living with Others"
    

df['Living_Status'] = df['Marital_Status'].apply(living_status)
    
```

```python
df.head()
```

```python
# Create "Age" field base on the following condition

df["Age"] = 2024 - df["Year_Birth"]
```

```python
# Create "Total_Campaigns_Accepted" field base on campaign accepted fields


df['Total_Campaigns_Accepted'] = df[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']].sum(axis=1)
```

```python
# Create "Average_Spend" field base on product and NumDealsPurchases fields


df['Average_Spend'] = (df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts',
                           'MntSweetProducts', 'MntGoldProds']].sum(axis=1)) / df['NumDealsPurchases']

```

```python
# Create "Spent field" base on the spent on products


df['Spent'] = df['MntWines']+df["MntWines"] +df['MntFruits']+ df['MntMeatProducts'] +df['MntFishProducts']+ df['MntSweetProducts']+ df['MntGoldProds']
```

```python
# Create "Is_Parent" to indicate parenthood status

df['Is_Parent'] = (df['Kidhome'] + df['Teenhome'] > 0).astype(int)
```

Creating total spending in the last 2 years,  average monthly visits to the company's website, ratio of online purchases to total purchases

```python
# Create "total spending" field based on spending in the last 2 years

df['total_spending'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']

```

```python
# Create average monthly visits to the company's website field

df['avg_web_visits'] = df['NumWebVisitsMonth'] / 12
```

```python
# Create ratio of online purchases to total purchases field

df['online_purchase_ratio'] = df['NumWebPurchases'] / (df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases'])

```

The heatmap visualizes the correlation between different campaign acceptance variables (AcceptedCmp1, AcceptedCmp2, AcceptedCmp3, and AcceptedCmp4), providing a comprehensive overview of their interrelationships. Each cell in the heatmap represents the correlation coefficient between two campaigns, indicating the strength and direction of their relationship.

```python
# Correlation matrix
correlation_matrix = df_no_outliers[["AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3", "AcceptedCmp4"]].corr()

# Heatmap for campaign correlations
fig_heatmap = px.imshow(correlation_matrix,
                         title="Campaign Acceptance Correlation",
                         color_continuous_scale='RdBu',
                         labels=dict(x="Campaigns", y="Campaigns", color="Correlation"),
                         x=correlation_matrix.columns,
                         y=correlation_matrix.columns)


fig_heatmap.show()
```
